```python
import os
import pandas as pd
```

```python
prefix = '_SELL_OUT_5_raw.csv'
score = 0.25
```

```python
def make_rate(version: str, prefix, score, mega_yn=False):
    data = pd.read_csv(os.path.join(version, version+prefix), encoding='cp949')
    temp = data[['cust_grp_nm', 'item_cd', 'acc_pred', 'mega_yn']].copy()
    
    if mega_yn == True:
        temp = temp[temp['mega_yn'] == 'Y']
    
    total_cnt = temp.groupby(by='cust_grp_nm')['item_cd'].count().reset_index()
    total_cnt = total_cnt.rename(columns={'item_cd': 'total'})
    
    filtered = temp[(temp['acc_pred'] >= (1-score)) & (temp['acc_pred'] <= (1+score))]
    filtered_cnt = filtered.groupby(by='cust_grp_nm')['item_cd'].count().reset_index()
    filtered_cnt = filtered_cnt.rename(columns={'item_cd': 'cover'})
    
    merged = pd.merge(total_cnt, filtered_cnt, on=['cust_grp_nm'])
    
    merged['rate'] = merged['cover'] / merged['total']
    
    return merged.T
```

```python
make_rate('20190311-20220306', prefix, score)
```
